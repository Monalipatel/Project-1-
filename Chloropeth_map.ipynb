```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
world = gpd.read_file("Resources/TM_WORLD_BORDERS-0.3.shp")
world
```

```python
df_2019 = pd.read_csv("Resources/2019.csv")
df_2019 = df_2019[["Country", "Happiness Score", "Economy (GDP per Capita)", "Family/Social Support", "Health (Life Expectancy)"]]
df_2019 = df_2019.rename(columns={"Country": "NAME"})
df_2019
```

```python
df_merge = pd.merge(world, df_2019, on="NAME")
df_merge
```

```python
new_df = df_merge[["NAME", "Happiness Score", "Economy (GDP per Capita)", "Family/Social Support", "Health (Life Expectancy)", "LAT", "LON", "geometry"]]
new_df
```

```python
happy_map = new_df.plot(figsize=(20,15), cmap = "Blues", column="Happiness Score", legend = True)
plt.savefig("Images/Happy_Map.png")
happy_map
```

```python
eco_map = new_df.plot(figsize=(20,15), cmap = "YlGn", column="Economy (GDP per Capita)", legend = True)
plt.savefig("Images/Eco_Map.png")
eco_map
```

```python
fam_map = new_df.plot(figsize=(20,15), cmap = "YlOrBr", column="Family/Social Support", legend = True)
plt.savefig("Images/Family_Map.png")
fam_map
```

```python
health_map = new_df.plot(figsize=(20,15), cmap = "Purples", column="Health (Life Expectancy)", legend = True)
plt.savefig("Images/Health_Map.png")
health_map
```
