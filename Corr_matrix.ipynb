```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
import scipy.stats as st
```

```python
df_2019 = pd.read_csv('Resources/2019.csv')
```

```python
df_2019.head()
```

```python
corr_matrix = df_2019[["Economy (GDP per Capita)", "Family/Social Support", "Health (Life Expectancy)", "Freedom", "Trust (Government Corruption)", "Generosity"]]
corr_matrix
```

```python
corr_matrix.corr()
```

```python
f = plt.figure(figsize=(15, 10))
plt.matshow(corr_matrix.corr(), fignum=f.number)
plt.xticks(range(corr_matrix.shape[1]), corr_matrix.columns, fontsize=10,rotation=90)
plt.yticks(range(corr_matrix.shape[1]), corr_matrix.columns, fontsize=10)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=10)

#plt.title('Correlation Matrix', fontsize=16);
png_filepath = "Images/Correlation Matrix.png"

plt.savefig(png_filepath)
```

# Analysis for Correlation Matrix


Happiness score essentially triple counts one factor that can be defined as either Health, GDP, or Family. 

If a country does well in any one these factors it will likely do well in the other 2, moving them up the list
